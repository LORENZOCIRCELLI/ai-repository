```python
operation_array_one = np.array(np.random.randint(0,99,size=(3,4)))
operation_array_two = np.array(np.random.randint(0,99,size=(3,4)))
```

```python
sum_array = operation_array_one + operation_array_two
print(sum_array)
difference_array = operation_array_one - operation_array_two
print(difference_array)
product_array = operation_array_one * operation_array_two
print(product_array)
division_array = operation_array_one / operation_array_two
print(division_array)
```

```python
exp_array = np.exp(operation_array_one)
print(exp_array)
log_array = np.log(operation_array_one)
print(log_array)
sqrt_array = np.sqrt(operation_array_one)
print(sqrt_array)
```

```python
maximum_array = np.maximum(operation_array_one, operation_array_two)
print(maximum_array)
minimum_array = np.minimum(operation_array_one, operation_array_two)
print(minimum_array)
```

```python
abs_array = np.abs(operation_array_one)
print(abs_array)
```
